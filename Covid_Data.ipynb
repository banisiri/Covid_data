```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pandas import Series, DataFrame
```

```python
df=pd.read_csv("Covid_Data.csv")
df
```

```python
df=df.drop(columns=["DATE_DIED","INTUBED","USMER","MEDICAL_UNIT","ICU"])
#df=df.iloc[:1000,:]
```

```python
pd.set_option('display.max_columns', None)
df.columns
```

```python
values=[99,98,97]
df = df[df.PNEUMONIA.isin(values) == False]
df = df[df.PREGNANT.isin(values) == False]
df = df[df.DIABETES.isin(values) == False]
df = df[df.COPD.isin(values) == False]
df = df[df.ASTHMA.isin(values) == False]
df = df[df.INMSUPR.isin(values) == False]
df = df[df.HIPERTENSION.isin(values) == False]
df = df[df.OTHER_DISEASE.isin(values) == False]
df = df[df.CARDIOVASCULAR.isin(values) == False]
df = df[df.OBESITY.isin(values) == False]
df = df[df.RENAL_CHRONIC.isin(values) == False]
df = df[df.TOBACCO.isin(values) == False]
```

```python
#df
#df.isnull().any()
#df.isnull().sum()
df=df.iloc[:1000,:]
```

```python
x=df.drop(columns=["CLASIFFICATION_FINAL"])
y=df.iloc[:,15:]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#from sklearn.preprocessing import StandardScaler
#sc=StandardScaler()
#x_train=sc.fit_transform(x_train)
#x_test=sc.fit_transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
params_knn={'n_neighbors': np.arange(1, 25)}
knn_gs=GridSearchCV(knn,params_knn,cv=5)
knn_gs.fit(x_train,y_train)
```

```python
knn_best=knn_gs.best_estimator_
print(knn_gs.best_params_)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier()

#create parameter dictionary to find the best among them

params_rf={'n_estimators':[50,100,200]}

rf_gs=GridSearchCV(rf,params_rf,cv=5)

rf_gs.fit(x_train,y_train)
```

```python
rf_best=rf_gs.best_estimator_
print(rf_gs.best_params_)
print(rf_gs.best_params_)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(max_iter=500)
log_reg.fit(x_train,y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc=GradientBoostingClassifier()
```

```python
params_gbc={'n_estimators':[50,100,200]}
gbc_gs=GridSearchCV(gbc,params_gbc,cv=5)
gbc_gs.fit(x_train,y_train)
gbc_best=gbc_gs.best_estimator_
print(gbc_gs.best_params_)
```

```python
knn_best.score(x_test,y_test),rf_best.score(x_test,y_test),log_reg.score(x_test,y_test),gbc_best.score(x_test,y_test)
```

```python
from sklearn.ensemble import VotingClassifier
estimators=[('knn',knn_best),('rf',rf_best),('log_reg',log_reg),('gbc',gbc_best)]
ensemble=VotingClassifier(estimators,voting='hard')
ensemble.fit(x_train,y_train)
ensemble.score(x_test,y_test)
```
